We can see that region 2, i.e **Europe & Central Asia** has had the highest aaverage CO2 emission overall.

```python
value_count = df['Region'].value_counts()
reverse_mapping = {v: k for k, v in area_mapping.items()}
value_count.index = value_count.index.map(reverse_mapping)
plt.figure(figsize=(10, 6))
value_count.plot(kind='bar')
plt.title('Emissions by Regions')
plt.xlabel('Region')
plt.ylabel('Emission(metric tons per cacpita)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Defining a function to give a line plot of the average CO2 emissions of a region over the years.**

```python
def plot_line_graph(df, region_name):
    year_columns = df.columns[2:]
    df_mean = df[year_columns].mean()
    plt.plot(year_columns, df_mean, marker='o')
    plt.title(f'Average CO2 Emissions in {region_name} Over Time')
    plt.xlabel('Year')
    plt.ylabel('Average CO2 Emissions (metric ton per capita)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
```

```python
plt.figure(figsize=(25, 12))
plt.subplot(3, 3, 1)
plot_line_graph(df1, 'East Asia & Pacific')
```

```python
plt.figure(figsize=(25, 12))
plt.subplot(3, 3, 2)
plot_line_graph(df2, 'Europe & Central Asia')
```

**Defining a function to give a line plot of the top 5 countries by average CO2 emissions in each region over the years and plotting the graphs.**

```python
def plot_top5_countries(df, region_name):
    year_columns = df.columns[2:]

    total_emissions = df[year_columns].sum(axis=1).sort_values(ascending=False)
    top_5_countries = total_emissions.head(5).index
    plt.figure(figsize=(10, 6))

    for country in top_5_countries:
        plt.plot(year_columns, df.loc[country, year_columns], label=country, marker='o')

    plt.title(f'Top 5 Countries by CO2 Emissions in {region_name} over the years')
    plt.xlabel('Year')
    plt.ylabel('CO2 Emissions (metric ton per capita)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.legend(loc='best', fontsize='small')
    plt.tight_layout()
    plt.show()
```

```python
plot_top5_countries(df1, 'East Asia & Pacific')
```

**Defining a function to give a heatmap of the CO2 emissions of all the countries in a region over the years and plotting the graphs.**

```python
def plot_emissions_heatmap(df, region_name):
    year_columns = df.columns[2:]
    plt.figure(figsize=(12, 8))

    sns.heatmap(df[year_columns], cmap='YlGnBu', linewidths=0.5, linecolor='gray')

    plt.title(f'CO2 Emissions Heatmap for Countries in {region_name} over the years', fontsize=14)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Country', fontsize=12)

    plt.tight_layout()
    plt.show()
```

The general trend of emissions in every region over the years is as follows:

* **East Asia & Pacific**
    
    The steady rise in CO₂ emissions in East Asia reflects rapid industrialization and urbanization, particularly in China, which heavily relies on coal to power its industries and energy needs. The region’s emissions surged after 2000, fueled by infrastructure projects and manufacturing. Efforts to invest in renewable energy are underway but haven’t yet offset emissions entirely due to high energy demands and reliance on coal​
* **Europe & Central Asia**

    The decline in CO₂ emissions in this region began in the 1990s after the Soviet Union's collapse, which caused a reduction in industrial activity and energy demand. Many countries modernized their energy systems, shifting to natural gas and adopting energy-efficient technologies. The European Union's strict climate policies and increased use of renewables further accelerated the reductions

* **Latin America & Carribbean**

    CO₂ emissions in Latin America have remained relatively stable, with occasional spikes. Key contributors include deforestation in the Amazon and urbanization, particularly in growing cities. However, a reliance on renewable energy like hydropower has kept emissions lower compared to other regions. Transportation and agricultural activities remain notable sources in emissions due to their unique socioeconomic challenges
* **Middle East & North Africa**

    CO₂ emissions in the MENA region rose sharply due to urbanization and reliance on oil and gas. Recently, some countries have started investing in renewables, such as large-scale solar projects, and improving energy efficiency. However, fossil fuel subsidies and economic dependence on oil exports still hinder significant emission reductions
* **North America**

    North America exhibit a declining trend in emissions, attributed to a transition from coal to cleaner energy sources like natural gas and renewables. Improved energy efficiency in industries and homes, coupled with a decline in heavy manufacturing, has contributed significantly. Stricter environmental policies and carbon regulations also play a critical role in reducing emissions​
* **South Asia**

    The steep increase in CO₂ emissions in South Asia, particularly after the early 2000s, aligns with its rapid economic growth and urbanization. India, for example, has become a major emitter due to its reliance on coal for energy production and growing transportation needs. However, South Asia's per capita emissions remain relatively low compared to developed regions, reflecting the ongoing energy access challenges in rural areas
* **Sub-Saharan Africa**

    While emissions in Sub-Saharan Africa have increased over the years, the overall levels remain modest. Economic development and urbanization have contributed to the gradual rise, with a focus on oil and biomass as energy sources. However, widespread investments in renewable energy and low per capita energy use keep emissions significantly lower compared to other regions​
